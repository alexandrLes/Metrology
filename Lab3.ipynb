```python
import numpy as np
from matplotlib.pyplot import (axes, axis, title, legend, figure,
                               xlabel, ylabel, xticks, yticks,
                               xscale, yscale, text, grid,
                               plot, scatter, subplot, errorbar, hist, polar,
                               contour, contourf, colorbar, clabel,
                               imshow)
from mpl_toolkits.mplot3d import Axes3D
from numpy import (linspace, logspace, zeros, ones, outer, meshgrid,
                   pi, sin, cos, sqrt, exp)
from numpy.random import normal
% matplotlib inline
```

```python
n = np.arange(1, 10)
Rp = np.arange(1, 10) / 1000
```

```python

g = subplot()
g.plot(n, Rp, label="Начальный участок графика статической характеристики ЦИП")
g.scatter(n, Rp, color='orange', s=40, marker='o')
g.grid()
g.legend()
```

```python
R = np.array([12, 22, 32.6, 42.5, 52.7, 62.6, 72.4, 83.3, 93.1]) / 10000
delta_R = Rp - R
```

```python
g = subplot()
g.plot(Rp, delta_R, label="График абсолютной основной погрешности ")
g.scatter(Rp, delta_R, color='orange', s=40, marker='o')
g.grid()
g.legend()
```
